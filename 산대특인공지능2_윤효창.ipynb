```python
# Question3 :
# 1)
import pandas as pd
data = pd.read_csv('data/Traffic_Accident_2017.csv',encoding = 'euc-kr')
car_acc = data[(data['사고유형_대분류'] == '차대사람')&(data['주야'] == '야간')]
car_acc2 = car_acc[['주야','요일']].groupby(by = "요일").count()
car_acc2
# 2)
x = car_acc2['주야'].values
x = car_acc2.index

y = car_acc2['주야']
y = y[['월','화','수','목','금','토','일']]
y

plt.ylim(0,200)
plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.bar(x,y)
```
