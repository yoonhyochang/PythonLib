```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### 선 그래프 그려보기


```python
#line plot
# 선 그래프 키워드 : plot
y = [2,4,6,8]
plt.plot(y)
plt.show()
```

```python
# 선 그래프 키워드 : plot
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y) #그래프 출력
```

##### 그래프 스타일 지정


```python
# 선 스타일
plt.plot(y,ls = "--")# ls : LINE STYLE 약자
plt.show()
```

```python
# 마커 포인트 : marker

plt.plot(y,marker = 'o')
plt.show()
```

```python
#  선 두께 :line width --> lw
plt.plot(y,marker = 'o', lw= 5)
plt.show
```

```python
# 선 색상 : line color --> c
plt.plot(y,marker="o",lw = 5, c= 'g')
plt.show
```

```python
#marker 의 컬러 지정하기
plt.plot(y,marker='o',c='g',markerfacecolor='m')
plt.show
```

```python
x = np.arange(7)# 0~6 범위를 만들면서 배열 데이터 생성 함수(numpy 모듈 불러오기)
y = [1,4,5,8,9,5,3]
```

```python
plt.plot(x,y,ls = "--",marker = 'o',c= 'g')
plt.show
```

```python
plt.plot(x,y,ls = ":",marker = 's',c= 'm',mfc='b')
plt.show
```

```python
#그래프 사이즈 지정
plt.figure(figsize= (5,3))
plt.plot(x,y,ls = ":",marker = 's',c= 'm',mfc='b')
plt.show
```

##### 그래프 옵션 설정


```python
# 그림 범위 지정
plt.plot(x,y,ls =':',c='r',marker='d',mfc='b')

plt.xlim(-2,10) #x축의 범위 제한
plt.ylim(-5,12) # y축의 범위 제한
plt.show
# 만약, 2~8의 사이의 값을 보고싶을 때(zoomin)-->y축 값의 범위 제한
```

```python
#틱 설정
plt.plot(x,y,ls =':',c='r',marker='d',mfc='b')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show
```

```python
# 눈금선 표시 :grid()

plt.plot(x,y,ls =':',c='r',marker='d',mfc='b')
plt.grid() # grid 함수 호출
plt.show
```

```python
# 여러개의 그래프 출력
x= [1,2,3,4]
y= [2,4,6,8]
z= [3,6,9,12]

plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend(loc ='best')# 범례표시 'best'==>'기본 값'
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('my graph') # 그래프 제목
plt.show()#x,y,z 모두 출력
```

##### 데이터 불러오기

```python
data = pd.read_csv('data/Traffic_Accident_2017.csv',encoding = 'euc-kr')
data
```

```python
data[:5]
data.head() # 맨위의 행에서 부터 5개 까지 출력
data.tail() # 맨아래 행에서 부터 5개 까지 출력

```

##### 요일별 사고 건수 count

```python
#1. 요일 컬럼 접근
data['요일']
```

```python
#2.중복없이 값 확인(월 ~ 일)
data['요일'].unique()
# 중복없이 값 확인하기 --> .unique

data.groupby(by=['요일']).count()
```

```python
#3. value_counts()
y = data['요일'].value_counts()
y 
```

```python

y = y[['월','화','수','목','금','토','일']]
y
```

```python
x = y.index  # x가  없으므로 요일 정보를 출력하기 위해 인덱스명 추출


plt.bar(x,y)
#옵션 설정
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 교통사고 건수')
plt.ylim(500,620)# Y축 범위 제한
plt.show()
```


##### 차대차 사건 중 사상자 수가 많은 발생지 시도를 알아보고 시각화하기
- 차대차사건, 사상자수 , 발생지 시도 ==> 접근

- 힌트
    - 사고유형 정보가 있는 컬럼에서 차대차인 행만 출력
    - 발생지 시도에 있는 각각의 데이터 회수 확인 ==> 확인용
    -'사상자 수', '발생지 시도' 컬럼 인덱싱
    - 발생지 시도별로 묶고 집계함수(총합)연결 후 시각화

```python
# 컬럼 확인
data.columns
```

```python
#컬럼 확인

car_acc = data[data['사고유형_대분류'] == '차대차']

car_acc
```

```python
car_acc2 = car_acc[['사상자수','발생지시도']].groupby(by='발생지시도').sum()
car_acc2
```

```python
x = car_acc2['사상자수'].values #컬럼은 한개지만 행이 여러개로 2차원 배열 데이터프레임에서 '사상자 수'
x
```

```python
x = car_acc2.index #발생지 시도에 접근하기
x
```

```python
y = car_acc2['사상자수']
y
```

```python
# 시각화
plt.bar(x,y)
plt.title('차대차 교통사고의 사상자 수')
plt.xlabel('지역')
plt.ylabel('사상자수')
plt.show()
```
