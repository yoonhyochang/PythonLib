- 영화 평점 데이터를 활용하여 각 사용자 별 평점 평균 구하기

```python
import numpy as np
```

```python
date = np.loadtxt('data/ratings.dat',delimiter = '::', dtype = np.int64)
date
```

```python
# 데이터 확인하기

# 배열의 크기 : shape
print('배열의 크기 : ',date.shape)
print('행 크기 : ',date.shape[0],'열크기 : ',date.shape[1] )

#배열의 차원 : ndim
print('배열의 차원 : ', date.ndim)

# 배열의 전체 요소 개수 : size
print('배열의 요소 개수 : ', date.size)
```

##### 전체 평점 평균 구하기

```python
# 평점데이터에 접근하고 평균함수 이용
total_mean = date[:,2].mean() #[행, 열] #모든행에 , 2번째 열 
total_mean
```

##### 1번 사용자 데이터에 접근하기

```python
# 0번째 열 - 사용자 아이디 열
date[date[:,0] == 1].size
```

```python
# 1 번 사용자의 평점의 평균 구해보기
a = date[date[:,0] == 1] # 1번 사용자의 평점 데이터에 접근
np.mean(a[:,2])
# date[date[:,0] == 1][:,2]
# date[date[:,0] == 1,2]
# 불리언 인덱싱은 행인덱싱 느낌으로 활용가능
```

```python
a[:,2]
```

```python
#영화 아이디 10번의  평점 평균 구하기
date[date[:,1]== 10,2].mean()
```

##### 중복없이 사용자 아이디 확인하기 : unique

```python
# 사용자 아이디 - 0번째 열
user_id =np.unique(date[:, 0])
np.unique(date[:, 0]).size
print(user_id)
print(user_id.size)
```

```python

# 반복문을 활용하여 각 사용자 아이디 출력해보고, 각 사용자별 평점의 평균 구하기
# 1. 저장을 위한 빈리스트
user_id_mean = []
#2 . 각사용자 아이디를 반복시키기
for user in user_id:
    user_mean = np.mean(date[date[:,0] == user,2])
    user_id_mean.append([user,user_mean])
user_id_mean
#3. for 내부에서 평점의 평균을 구하기
    
#     user_mean = date[date[:,0] == user,2].mean()
   
#4 . for 내부에서 [사용자 아이디 ,평점평균 ]형태로 저장
#5. for 외부에서 저장한 리스트 출력하기

    
   


    
```

```python
# 각 사용자별 평균 평균이 4점 이상인 아이디 구하기(불리언 인덱싱을 하기위해 리스트-->배열로 전환 필요)
user_id_mean2 = np.array(user_id_mean, dtype = np.int64)

user_id_mean3 = user_id_mean2[user_id_mean2[: ,1] >= 4][ : ,0]

print(user_id_mean3.size)

```
