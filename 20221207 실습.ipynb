```python
import numpy as np
```

```python
data = np.loadtxt('data/height_weight.txt', delimiter=',')
data
```

- Numpy 연산, 인덱싱 기법을 이용한 BMI 지수 구하기
1. BMI 연산
2. boolean 인덱싱 ---> 과체중 데이터 접근

```python
#배열 데이터 확인
data.shape #2행 10열(2,10)
#데이터 타임
data.dtype#flot64
#차원확인
data.ndim #2차원
data[0]
```

```python
#1. 키 변수에 키 배열 데이터를 담아주기
h = data[0]
print(h)
#2. 몸무게 변수에 몸무게 배열 데이터를 담아주기
w  = data[1]
print(w)
#3. 키 데이터 단위 변경하기 cm --> m(요소 연산)
m_h = h*0.01
print(m_h)
#4. 배열간 연산을 통해서 BMI 지수 구하기 (공식적용)
bmi = w/(m_h**2)
print(bmi)

#5. boolean 안덱싱을 활용하여 과체중(23<= bmi <=25)이상이 데이터에 접근
# 비트연산자 : & -and(두 값 모두 True 일때 True 반환)
#               | - or(두 값 중 하나만  True여도 True 반환)
bmi[(bmi >= 23) & (bmi <= 25)]
    
#6. 과체중인 사람수(데이터 갯수)
bmi[(bmi >= 23) & (bmi <= 25)].size
```
